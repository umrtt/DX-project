```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, BatchNormalization
```

```python
base_model = tf.keras.applications.EfficientNetB0(input_shape=(224,224,3),include_top=False,weights='imagenet')
model = keras.Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(256,activation='relu'),
    BatchNormalization(),
    Dense(128,activation='relu'),
    Dense(1,activation='sigmoid')]
)
model.summary()
keras.utils.plot_model(model,to_file='my_model.png',
                      show_shapes=True)
```

```python
from graphviz import Digraph

# Create a new directed graph
dot = Digraph()

# Add nodes for each layer
dot.node('Input', 'Input Layer')
dot.node('Dense1', 'Dense Layer 1\n(64 units, ReLU)')
dot.node('Dense2', 'Dense Layer 2\n(10 units, Softmax)')

# Add edges to represent the flow of data
dot.edge('Input', 'Dense1')
dot.edge('Dense1', 'Dense2')

# Render the graph to a file
dot.render('model_architecture', format='png', cleanup=True)

# Display the graph
dot.view()
```
